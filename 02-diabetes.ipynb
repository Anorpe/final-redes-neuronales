## Lectura de Datos

Se hace una lectura simple de la base proporcionada para el desarrollo de este ejercicio.

```python
import pandas as pd

data = pd.read_csv('datos/diabetes.csv')
data
```

# Regresión Lineal

## Procesamiento de Datos

Para el procesamiento de datos se procede a estandarizar los datos con la funcion RobustScaler de la libreria de Sklearn, la cual es una estandarización menos sensible a datos con varianzas altas.

```python
from sklearn.preprocessing import RobustScaler
dataux = data
columnas = data.columns.values
robust = RobustScaler().fit(dataux)
dataux = robust.transform(dataux)
dataux = pd.DataFrame(dataux,columns=columnas)
print(dataux)
X = dataux.drop('Y',axis = 1)
X_name = X.columns.values
y = dataux['Y']

print(X)
print(y)
print("tamaño X",X.shape)
print("tamaño de Y", y.shape)
```

## Selección de Caracteristicas

Se procede a realizar una busqueda de caracteristicas mas significativos con el metodo RFECV proporcionado por la libreria Sklearn el cual utiliza como base para su busqueda un estimador SGDregressor de la mismas libreria, el cual es el mismo que vamos a utilizar para la creacion del modelo de regresión lineal.

```python
from sklearn.feature_selection import RFECV
from sklearn import linear_model
estimator = linear_model.SGDRegressor(max_iter=1000, tol=1e-3)
selector = RFECV(estimator, step=1, cv=5)
selector = selector.fit(X, y)
indices = selector.get_support(indices = True)
carac = X_name[indices]
print(carac)

```

```python
X = X[carac].values
X
```

## Ajuste de Parametros

Para la busqueda de los mejores parametros se utiliza la funcion GridSearchCV de Sklearn, la cual dado un dominio de los parametros hace una busqueda exhaustiva de la combinación de parametros que me genere el mejor valor de una metrica escogida usando validación cruzada, la cual en este caso es $ R^2$

```python
from sklearn.model_selection import GridSearchCV
parametros = {
    'max_iter':[10000],
    'tol':[0.0001,0.001,0.01,0.1,1,10,100,1000],
    'loss':['squared_loss', 'huber', 'epsilon_insensitive', 'squared_epsilon_insensitive'],
    'penalty':['none', 'l2', 'l1', 'elasticnet'],
    'alpha':[0.001,0.01,0.1,1,10],
    'learning_rate':['constant', 'optimal', 'invscaling', 'adaptive']
}
linear = linear_model.SGDRegressor()
clf = GridSearchCV(linear, parametros,cv=5,scoring='r2')
clf.fit(X,y)
print("Mejores Parametros Encontrados")
print(clf.best_params_)
```

# Entrenamiento y validación del modelo

Con base en los mejores parametros encontrados procedemos a entrenar el modelo y validarlo con las metricas $R^2$ y $MSE$ usando la tecnica de validacion cruzada.

```python
from sklearn import linear_model
import numpy as np
linear = linear_model.SGDRegressor(alpha= clf.best_params_['alpha'], 
                                   learning_rate= clf.best_params_['learning_rate'], 
                                   loss=clf.best_params_['loss'], 
                                   max_iter = clf.best_params_['max_iter'], 
                                   penalty = clf.best_params_['penalty'], 
                                   tol=clf.best_params_['tol'])

from sklearn.model_selection import cross_val_score
scores = cross_val_score(linear, X, y, cv=5,scoring = 'neg_mean_squared_error')
print("MSE Promedio",np.mean(abs(scores)))
scores = cross_val_score(linear, X, y, cv=5,scoring = 'r2')
print("R2",np.mean(scores))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
linear = linear.fit(X_train,y_train)
predict = linear.predict(X_test)
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(range(0,30),y[len(y_test)-30:len(y_test)],color='black')
plt.plot(range(0,30),predict[len(predict)-30:len(predict)],color='red')
```

## Ajuste de Parametros

```python
from sklearn.model_selection import GridSearchCV
import sklearn.neural_network

parametros = {"hidden_layer_sizes": [(1,),(6,),(12,),(30,),(50,),(100,),(1000,),(1,2),(6,2),(12,2),(30,2),(50,2),(100,2),(1000,2)],
              "activation": ["identity", "logistic", "tanh", "relu"], 
              "solver": ["lbfgs", "sgd", "adam"], 
              "alpha": [0.00005,0.0005],
               'learning_rate':['constant', 'invscaling', 'adaptive'],
            #'max_iter':[200,1000],
            #'tol':[0.00001,0.0001,0.001,0.01,0.1,1,10],
            #'momentum':[0.1,0.2,0.4,0.6,0.8,1]}
             }
red = sklearn.neural_network.MLPRegressor()
clf = GridSearchCV(red, parametros,cv=5,scoring='r2')
clf.fit(X,y)
print("Mejores Parametros Encontrados")
print(clf.best_params_)
```

## Entrenamiento y Validación

Con base en los mejores parametros encontrados procedemos a entrenar el modelo y validarlo con las metricas $R^2$ y $MSE$ usando la tecnica de validacion cruzada.

```python
from sklearn import linear_model
import numpy as np
neural_net =  sklearn.neural_network.MLPRegressor(
                 hidden_layer_sizes = (100,),  # Una capa oculta con una neurona
                activation = 'tanh',    #  {‘identity’, ‘logistic’, ‘tanh’, ‘relu’}
                 solver = 'adam',             #  {‘lbfgs’, ‘sgd’, ‘adam’}
                 alpha = 5e-05,                #
                # learning_rate_init = 0.1,   # Valor de la tasa de aprendizaje
                 learning_rate = 'invscaling', # La tasa no se adapta automáticamente
                # verbose = False,            # Reporte del proceso de optimización
                # shuffle = True,             #
                # tol = 1e-8,                 #
                # max_iter = 25000,           # Número máximo de iteraciones
                # momentum = 0.0,             #
                # nesterovs_momentum = False)
)#

from sklearn.model_selection import cross_val_score
scores = cross_val_score(neural_net, X, y, cv=5,scoring = 'neg_mean_squared_error')
print("MSE Promedio",np.mean(abs(scores)))
scores = cross_val_score(neural_net, X, y, cv=5,scoring = 'r2')
print("R2",np.mean(scores))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
neural_net = neural_net.fit(X_train,y_train)
predict = neural_net.predict(X_test)
import matplotlib.pyplot as plt
plt.plot(range(0,30),y[len(y_test)-30:len(y_test)],color='black')
plt.plot(range(0,30),predict[len(predict)-30:len(predict)],color='red')
```
